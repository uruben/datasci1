```python

import seaborn as sns

sns.set(context='talk')
```

```python

fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(12,6))


football_df['xG'].plot.hist(ax=axes[0], bins=50, alpha=0.5, color='orange')

axes[0].set_title('Expected Goals')


football_df['Matches_Played'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='red')
football_df['Goals'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='blue')
axes[1].set_title('matches played (red) and goals (blue)')

football_df['Shots'].plot.hist(ax=axes[2], bins=50, alpha=0.5, color='green')
axes[2].set_title('Shots')

```

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

football_df['League'].value_counts().plot(kind='pie', ax=axes[0])
football_df['Country'].value_counts().plot(kind='pie', ax=axes[1])


axes[0].set_ylabel('')
axes[1].set_ylabel('')

fig.suptitle('Pie charts of leagues and countries')


plt.tight_layout()
```

```python
goals_by_country = football_df.groupby('Country')['Goals'].sum().reset_index()
goals_by_country = goals_by_country.sort_values(by='Goals', ascending=False) # sort by descending order to look pretty

plt.figure(figsize=(10, 6))
plt.bar(goals_by_country['Country'], goals_by_country['Goals'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Nation')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
football_df= pd.read_csv("Football.csv")
```

```python
summary_stats = football_df[['Matches_Played', 'Goals', 'Shots']].describe()
```

```python
print(summary_stats)
```

```python
# comapring english and french players

english_players = football_df[football_df['Country'] == 'England']
french_players = football_df[football_df['Country'] == 'France']

total_goals_english = english_players['Goals'].sum()
total_goals_french = french_players['Goals'].sum()

print(f"Total goals scored by English players: {total_goals_english}")
print(f"Total goals scored by French players: {total_goals_french}")
```

```python
# visual 

fig, ax = plt.subplots()
ax.bar(['English', 'French'], [total_goals_english, total_goals_french], color=['red', 'blue'])
ax.set_ylabel('Total Goals')
ax.set_title('Comparison of Total Goals: English vs. French Players')
plt.show()
```

```python
# now find matches played for context

english_players = football_df[football_df['Country'] == 'England']
french_players = football_df[football_df['Country'] == 'France']

matches_played_english = english_players['Matches_Played'].sum()
matches_played_french = french_players['Matches_Played'].sum()

print("Value counts for Matches_Played - English players:")
print(matches_played_english)

print("\nValue counts for Matches_Played - French players:")
print(matches_played_french)
```
